## 绘制 Dual MBConv VAE 训练损失和Dice系数变化图，观察模型停止时刻，epoch=200
配置文件：```./train_configuration/test.json```

````python
matches_data = extract_data_from_log(
    log_path="../training_record/log/train_dual_MBConv_VAE.log",
    logstamp='2024-10-22 22:29:50 seg oasis4',
    fields=['EPOCH', 'train_total_loss', 'train_seg_loss', 'train_recon_loss','train_KL_loss','train_contras_loss','train_dice',
            'val_total_loss', 'val_seg_loss', 'val_recon_loss', 'val_KL_loss', 'val_contras_loss', 'val_dice']
)
````

````python
draw_line_graph(
    x=matches_data['EPOCH'],
    y_list=[
        matches_data['train_total_loss'],
        matches_data['train_seg_loss'],
        matches_data['train_recon_loss'],
        matches_data['train_KL_loss'],
        matches_data['train_contras_loss'],
        matches_data['train_dice'],
    ],
    label_list=[
        'train_total_loss',
        'train_seg_loss',
        'train_recon_loss',
        'train_KL_loss',
        'train_contras_loss',
        'train_dice',
    ],
    ylabel='Value',
    color_list=['orange','yellow', 'g', 'b', 'pink','r'],
    title='Dual_MBConv_VAE Train Loss',
    graph_name='Dual_MBConv_VAE Train Loss with 200 epoch'
)
````

````python
draw_line_graph(
    x=matches_data['EPOCH'],
    y_list=[
        matches_data['val_total_loss'],
        matches_data['val_seg_loss'],
        matches_data['val_recon_loss'],
        matches_data['val_KL_loss'],
        matches_data['val_contras_loss'],
        matches_data['val_dice'],
    ],
    label_list=[
        'val_total_loss',
        'val_seg_loss',
        'val_recon_loss',
        'val_KL_loss',
        'val_contras_loss',
        'val_dice',
    ],
    ylabel='Value',
    color_list=['orange', 'yellow','g', 'b', 'pink','r'],
    title='Dual_MBConv_VAE Validation Loss',
    graph_name='Dual_MBConv_VAE Validation Loss with 200 epoch'
)
````
